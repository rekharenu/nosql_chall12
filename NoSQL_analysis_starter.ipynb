### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {"scores.Hygiene":20}

# Use count_documents to display the number of documents in the result
number_results= establishments.count_documents(query)
print(f"number_results")
print(f"the establishments with hygine score 20 is: {number_results}")

# Display the first document in the results using pprint
results_first_document= establishments.find(query)
print("")
print("the first document:")

pprint(results_first_document[0])
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
london_query = {"localAuthorityName":{'$regex':"London"},"RatingValue":{'$gte':4}} 
results= establishments.find(london_query)


# Use count_documents to display the number of documents in the result
display_number=establishments.count_documents(london_query)
print(f"the count documents are{display_number}documents for establishments within 'London' with a 'RatingValue'>=4")

# Display the first document in the results using pprint

print("")
print("the first document is:")
pprint(display_number[0])
pprint(results[0])
```

```python
# Convert the result to a Pandas DataFrame
london_rating = pd.json_normalize(results)

# Display the number of rows in the DataFrame
london_rating.to_csv("csv_output/london_ratig.csv",index=False,header=True)

# Display the first 10 rows of the DataFrame
print(f"the number of rows :{len(london_rating)}")
print("")
london_rating.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score
lat_long= establishments.find_one({"BusinessName":"Penang Flavours"},{"geocode":1})


degree_search = 0.01
latitude = lat_long['geocode']['latitude']

longitude = lat_long['geocode']['longitude']

query ={"geocode.latitude":{'$gte':(latitude - degree_serch),'$lte':(latitude + degree_serch)},
        "geocode.longitude":{'$gte':(longitude - degree_serch),'$lte':(longitude + degree_serch)},
        "RatingValue":5}

sort =[("scores.Hygiene",1)]
limit = 5

# Print the results
results= list (establishments.find(query).sort(sort).limit(limit))
for rts in results:
    pprint(rts)
```

```python
# Convert result to Pandas DataFrame
results_df= pd.json_normalize(results)
results_df.to_csv=("csv_output/neray")
```
